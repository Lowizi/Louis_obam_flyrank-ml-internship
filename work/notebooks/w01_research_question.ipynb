**Lane: Refresh / Content Opportunity Scoring**

## 2. The question: decision, action, cost of a wrong call

**Decision:** Which page should an editor review and potentially refresh next?

**Who acts:** A content editor or SEO specialist with limited weekly capacity (e.g. 20-50 pages per week).

**Cost of a wrong call:**
- **False positive** (flagging a page that doesn't need work): wasted editor hours, reduced trust in the queue.
- **False negative** (missing a truly declining page): continued traffic loss, delayed recovery, missed revenue.

**Why ML helps:** The signals are real but noisy — position, CTR, engagement, age, and trend all interact in ways too tangled to capture with a hand-written formula. The pipeline demonstrates this: the hand-rule scores 0.240 Precision@50, while the model scores 0.740. A learned model captures interactions (e.g. "old content is risky only if it has impressions") that are tedious to write and maintain as rules.

## 3. Quick look at the data (2-3 real numbers)

The numbers below show three things that make this lane worth pursuing: (1) the problem is real and large, (2) a model already beats a hand-rule by 3x, and (3) there are thousands of candidates with real demand.

```python
import pandas as pd, json
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

total = len(df)
declining = (df["trend_direction"] == "down").sum()
decl_pct = declining / total * 100
print(f"1. {declining:,} of {total:,} pages are declining ({decl_pct:.1f}%) \u2014 the problem is real and large.")

high_imp = (df["impressions_90d"] >= 100).sum()
decl_high = ((df["trend_direction"] == "down") & (df["impressions_90d"] >= 100)).sum()
print(f"2. {decl_high:,} declining pages still have 100+ impressions \u2014 thousands of actionable candidates.")

res = json.load(open("outputs/model_results.json"))
base = res["baseline"]["baseline_precision_at_50"]
rf = res["models"]["random_forest"]["precision_at_50"]
print(f"3. Random forest Precision@50: {rf:.3f} vs baseline: {base:.3f} \u2014 {rf/base:.1f}x lift. Real signal exists.")
```
